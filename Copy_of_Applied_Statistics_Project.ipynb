##Task 2 - Descriptive Statistics üìä

- **Calculate measures of central tendency (mean, median, mode) and measures of dispersion (variance, standard deviation) for key variables. Identify and handle outliers if necessary.**


```python
import pandas as pd
import numpy as np
from scipy import stats


def analyze_data(df, column):
    print(f"--- Analysis for: {column} ---")

    # Calculate Central Tendency
    mean_val = df[column].mean()
    median_val = df[column].median()
    mode_val = df[column].mode()[0]

    # Calculate Dispersion
    variance_val = df[column].var()
    std_dev_val = df[column].std()

    print(f"Mean: {mean_val:.2f} | Median: {median_val} | Mode: {mode_val}")
    print(f"Variance: {variance_val:.2f} | Std Dev: {std_dev_val:.2f}")

    # 2. Identify Outliers using IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column].tolist()
    print(f"Detected Outliers: {outliers}")

    # 3. Handle Outliers (Capping/Winsorizing)
    # We replace outliers with the upper/lower bounds to keep data size consistent
    df_clean = df.copy()
    df_clean[column] = np.where(df_clean[column] > upper_bound, upper_bound,
                       np.where(df_clean[column] < lower_bound, lower_bound, df_clean[column]))

    print(f"New Mean (after capping): {df_clean[column].mean():.2f}\n")
    return df_clean

# Run analysis for each column
for col in df.columns:
    df = analyze_data(df, col)
```

#**Binomial Distribution**

```python
from scipy.stats import binom

#Parameters
n = 10    # trials (e.g., number of lightbulbs)
p = 0.05  # probability of success (e.g., defect rate)
k = 2     # target successes

#Calculate Probability P(X = 2)
prob = binom.pmf(k, n, p)

#Calculate Expected Value (Mean)
mean, var = binom.stats(n, p, moments='mv')

print(f"Probability of exactly {k} defects: {prob:.4f}")
print(f"Expected number of defects: {mean}")
```

#**Normal Distribution**

```python
from scipy.stats import norm

# Parameters
mu = 175    # Mean (e.g., average height in cm)
sigma = 7   # Standard Deviation
x = 189     # Target value

#Calculate Probability P(X > 189)
# norm.cdf(x, mu, sigma) gives P(X <= x)
prob_less_than = norm.cdf(x, mu, sigma)
prob_greater_than = 1 - prob_less_than

#Find Expected Value (Directly mu)
mean = norm.mean(loc=mu, scale=sigma)

print(f"Probability height is > {x}cm: {prob_greater_than:.4f}")
print(f"Expected value: {mean}cm")
```
